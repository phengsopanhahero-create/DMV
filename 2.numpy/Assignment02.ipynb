```python
import numpy as np
```

```python
# 1. Create array 1 to 10
a1 = np.arange(1, 11)
print(a1)
```

```python
# 2. Create array of zeros (10)
a2 = np.zeros(10)
print(a2)
```

```python
# 3. Create array of ones (10)
a3 = np.ones(10)
print(a3)
```

```python
# 4. Create array of even numbers 2–20
a4 = np.arange(2, 21, 2)
print(a4) 
```

```python
# 5. Create array of 10 random numbers (0–1)
a5 = np.random.rand(10)
print(a5)
```

```python
# 6. First 5 elements
a6 = a1[:5]
print(a6)
```

```python
# 7. Last 5 elements
a7 = a1[-5:]
print(a7)
```

```python
# 8. Reverse array
a8 = a1[::-1]
print(a8)
```

```python
# 9. Elements from index 2 to 7
a9 = a1[2:8]
print(a9)
```

```python
# 10. Even numbers from 1–50
a10 = np.arange(1, 51)
even = a10[a10 % 2 == 0]
print(a10)
```

```python
# 11. 4x5 random (0–10), filter rows
a11 = np.round(np.random.uniform(0, 10, (4, 5)), 1)
filtered = a11[(a11[:,1] > 2.0) & (a11[:,2] < 5.0)]
print(a11)
print(filtered)
```

```python
# 12. Replace numbers divisible by 2 or 3 but not both → -1
a12 = np.arange(1, 21)
cond = ((a12 % 2 == 0) ^ (a12 % 3 == 0))
a12[cond] = -1
print(a12)
print(cond)
print(a12[cond])
```

```python
# 13. 1D → 2D (5 rows)
a13 = np.arange(20).reshape(5, 4)
print(a13)
```

```python
# 14. 2D → 1D
a14 = a13.flatten()
print(a14)
```

```python
# 15. sum & mean
a15 = np.arange(1, 11)
sum_val = np.sum(a15)
mean_val = np.mean(a15)
print(a15)
print(sum_val)
print(mean_val)
```

```python
# 16. sum of second column
a16 = np.random.randint(1, 10, (4, 5))
col_sum = np.sum(a16[:,1])
print(a16)
print(col_sum)
```

```python
# 17. mean of first 3 rows
mean_rows = np.mean(a16[:3])
print(mean_rows)
```

```python
# 18. min per row & column
min_row = np.min(a16, axis=1)
min_col = np.min(a16, axis=0)
print(min_row)
print(min_col)
```

```python
# 19. Remove duplicates
a19 = np.array([1,2,2,3,4,4,5])
unique = np.unique(a19)
print(a19)
print(unique)
```

```python
# 20. Sort ascending & descending
a20 = np.random.randint(10, 21, 10)
asc = np.sort(a20)
desc = asc[::-1]
print(a20)
print(asc)
print(desc)
```

```python
# 21. Merge arrays
a21_1 = np.array([1,2,3])
a21_2 = np.array([4,5,6])
merged = np.concatenate((a21_1, a21_2))
print(merged)
```

```python
# 22. Common elements
a22_1 = np.random.randint(10, 21, 10)
a22_2 = np.random.randint(10, 21, 10)
common = np.intersect1d(a22_1, a22_2)
print(a22_1)
print(a22_2)
print(common)
```

```python
# 23. Remove 1,2,3,4
a23 = np.random.randint(1, 11, 10)
filtered = a23[~np.isin(a23, [1,2,3,4])]
print(a23)
print(filtered)
```

```python
# 24. Index of max value
a24 = np.arange(0.0, 5.1, 0.5)
max_index = np.argmax(a24)
print(a24)
print(max_index)
```

```python
# 25. First even > 5
a25 = np.array([1,3,7,8,4,10])
idx = np.where((a25 % 2 == 0) & (a25 > 5))[0][0]
print(a25)
print(idx)
```

```python
# 26. Add new column = col2 * col3
a26 = np.random.randint(1, 10, (4,5))
new_col = (a26[:,1] * a26[:,2]).reshape(-1,1)
result = np.hstack((a26, new_col))
print(a26)
print(new_col)
print(result)
```
